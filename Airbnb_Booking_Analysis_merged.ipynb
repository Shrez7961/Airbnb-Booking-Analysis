# **What are some frequently used  name which is searched by customer?**

To answer this question we are going to create a word cloud of data in name column which will show the presence of some attractive words that can help customers to get a glimpse of rooms/properties they are looking for and it will also attract customers.

For this purpose we must import word cloud.

```python
#word cloud
from wordcloud import WordCloud
```

Our name data may include many stop words so we have to strip it out. For this we have to import nltk

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
def stop_word_removal(data):
  stop = stopwords.words('english')
  data.name = data.name.str.replace(",", "").str.lower()
```

```python
Frequently_used_words = stop_word_removal(Airbnb_data)
Frequently_used_words = " ".join(str(each) for each in Airbnb_data.name)
```

```python
wordcloud = WordCloud(width=1980, height=1080, colormap="Paired_r",max_words=100, background_color="white").generate(Frequently_used_words)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## **What can we learn from price predictions?**

```python
#Plotting price over neighbourhood groups
plt.figure(figsize=(10,6))
plt.title("Neighbourhood Group vs price")
sns.barplot(x=Airbnb_data.neighbourhood_group, y=Airbnb_data.price,palette="muted") 
plt.show()
```

Again manhattan is in the top. It shows that manhattan is quite expensive compared to others. Lets distribute the price in categories as expensive affordable and cheap and check the distribution over categories for all the properties

**Here we consider the price less than or equal to 80 as cheap,more than 80 but less than 500 as affordable and more than 500 as expensive. **

```python
Airbnb_data['price_category']=Airbnb_data.apply((lambda x: 'cheap' if x['price']<=80 else 'affordable' if x['price']>=80 and x['price']<=500 else 'Expensive' ),axis=1)
Airbnb_data.head()
```

```python
#Lets check unique price category counts
Price_category=Airbnb_data.groupby('price_category')['id'].count().reset_index(name='number')
Price_category
```

The following pie chart shows the distribution of all the price categories in the given data.

```python
# Defining colors for the pie chart
colors = ['yellow', 'green', 'steelblue']
  
# Define the ratio of gap of each fragment in a tuple
explode = (0.05, 0.05, 0.05)
  
# Plotting the pie chart for above dataframe
Price_category.groupby(['price_category']).sum().plot(
    kind='pie', y='number', autopct='%1.0f%%',
  colors=colors, explode=explode)
```

The above pie chart shows the price category distribution in the dataset. Least people prefer the expensive category, instead maximum people prefer affordable category followed by cheap category.

```python
 #Grouping neighbourhood group with price category
 price_category_counts= Airbnb_data.groupby(['neighbourhood_group','price_category'])['id'].count().reset_index(name='number')
 print(price_category_counts)
```

```python
#Plot of neighbourhood group with price category
plt.figure(figsize = (15,5))
z = sns.barplot(y= 'number', x= 'neighbourhood_group',hue='price_category',data=price_category_counts)
z.set_title('RELATIONSHIP BETWEEN NEIGHBOTHOOD GROUPS AND PRICE CATEGORY')
z.set_xlabel('Neighbourhood Group')
z.set_ylabel("Price")
plt.show()
```

The above bar graph shows the relationship between the neighbourhood groups and the price category. Least people prefer the expensive category, instead maximum people prefer affordable category followed by cheap category in all the neighbourhood groups except in case of Bronx and Queens where the relationship is reverse.

# **Which hosts are the busiest and why?**

# **What can we learn from reviews predictions with respect to neighbourhood groups?**

```python
#Reviews analysis over neighbourhood and room type
reviews_prediction=Airbnb_data.groupby(['neighbourhood_group','room_type'])['number_of_reviews'].mean().reset_index(name='room_reviews')
reviews_prediction
```

# **Is there any noticeable difference of traffic among different areas and what could be the reason for it?**

```python
#Plotting the difference of traffic among different areas 
plt.figure(figsize=(10,6))
plt.title("Density and distribution of prices of neighberhood_group")
subset_df=Airbnb_data[Airbnb_data.price < 500]
#using violinplot to showcase density and distribtuion of prices 
viz_2=sns.violinplot(data=subset_df, x='neighbourhood_group', y='price')
```

* With a violin plot we can definitely observe a couple of things about distribution and density of prices for Airbnb in NYC Groups.

* This distribution and density of prices were completely expected; for example, as it is no secret that Manhattan is one of the most expensive places in the world to live in, where Bronx on other hand appears to have lower standards of living.
