# **4)Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?**

```python
df1['crew'].shape
```

```python
#Creating a function which returns the listr of producers and directors for each movie
def get_producers_and_directors(crew_list):
  try:
    #Initialize two lists for producers and directors
    producer_list=[]
    director_list=[]

    #iterate over the list elements
    for elem in crew_list:
      if elem['job']=='Producer':
        producer_list.append(elem['name'])
      if elem['job']=='Director':
        director_list.append(elem['name'])

    # row['producers'] = producer_list
    # row['directors'] = director_list  
    return producer_list, director_list
  except:
    return None

```

```python
#create no null values crew column
df1 = df[~df['crew'].isna()]
```

```python
#Convert no_null_crew_df from string to list
df1.loc[~df1['crew'].isnull(),'crew'] = df1.loc[~df1['crew'].isnull(),'crew'] .apply(convert_str_to_list)
```

```python
df1.iloc[1]['crew']
```

```python
get_producers_and_directors(df1.iloc[1]['crew'])
```

```python
df1[["producers","directors"]] = df1.apply(lambda row: get_producers_and_directors(row['crew']), axis=1, result_type='expand')
```

```python
df1.head()
```

```python
df1['profit']=df1['revenue']-df1['budget']
```

```python
df1['ROI']=100 * df1['profit']/df1['budget']
```
