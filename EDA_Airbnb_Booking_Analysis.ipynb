# **Exploratory Data Analysis**
What is Exploratory Data Analysis (EDA) ? 

EDA is a phenomenon under data analysis used for gaining a better understanding of data aspects like: – main features of data – variables and relationships that hold between them – identifying which variables are important for our problem We shall look at various exploratory data analysis methods like:

*  Descriptive Statistics, which is a way of giving a brief overview of the dataset we are dealing with, including some measures and features of the sample Grouping data [Basic grouping with group by] 

*  ANOVA, Analysis Of Variance, which is a computational method to divide variations in an observations set into different components. 
*  Correlation and correlation methods

**TYPES OF EXPLORATORY DATA ANALYSIS**:



*   Univariate (Graphical & Non-graphical)
*   Multivariate (Graphical & Non-graphical)

Further, we are going to create a word cloud of data in name column which will show the presence of some attractive words that can help customers to get a glimpse of rooms/properties they are looking for and it will also attract customers.

For this purpose we must import word cloud.

```python
#word cloud
from wordcloud import WordCloud
# ImageColorGenerator
# sns.set_theme(style="ticks", color_codes=True)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words('english'))
```

```python
def stop_word_removal(data):
  stop = stopwords.words('english')
  df.name = df.name.str.replace(",", "").str.lower()
  df.name = df.name.apply(lambda x: ' '.join([item for item in x.split() if item not in stop]))
```

```python
Frequently_used_words = stop_word_removal(df)
Frequently_used_words = " ".join(str(each) for each in df.name)
```

```python
wordcloud = WordCloud(width=600, height=480, colormap="Oranges_r",max_words=100, background_color="white").generate(Frequently_used_words)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
k=df.groupby(['neighbourhood_group','room_type'])['number_of_reviews'].mean().reset_index(name='room_reviews')
k
```

```python
plt.figure(figsize = (10,7))
ax_11 = sns.barplot(x='neighbourhood_group',y='room_reviews', hue ='room_type',data=k)
ax_11.set_title(' Room Type vs neighbourhood_group')
ax_11.set_xlabel('Room Type')
ax_11.set_ylabel("number_of_reviews")
plt.show()
```

```python
av=df.groupby(['neighbourhood_group','room_type'])['availability_365'].mean().reset_index(name='room availability')
av
```

```python
plt.figure(figsize = (10,5))
ax_11 = sns.barplot(x='neighbourhood_group',y='room availability', hue ='room_type',color = 'slategrey',data=av)
ax_11.set_title(' Room Type vs neighbourhood_group vs availability_365')
ax_11.set_xlabel('Room Type in  neighbourhood_groups')
ax_11.set_ylabel("Average availability 365 days")
ax_11.set_facecolor("#CDC673")
plt.show()
```
